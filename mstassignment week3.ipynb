```python

import mysql.connector
mydb=mysql.connector.connect(host="localhost",user="root",password="",database="EMS")

def searchemployee(keyword):
    cursor=mydb.cursor()
    query="select*from EMPLOYEE where emp_name LIKE'%{}'".format(keyword)
    cursor.execute(query)
    result=cursor.fetchall()
    if result:
        print("employee details")
        print("................")
        for row in result:
            print("employee id:",row[0])
            print("employee name:",row[1])
            print("employee salary:",row[2])
            print("employee department:",row[3])
            print("................")
    else:
        print("no matching employee found")

def insert_employee(idd,name,sal,dpt,mail):
    cursor=mydb.cursor()
    query="INSERT INTO EMPLOYEE VALUES(%s,%s,%s,%s,%s)"
    value=(idd,name,sal,dpt,mail)
    cursor.execute(query,value)
    cursor.fetchall()
    mydb.commit();
    print("....................")
    print("successfully inserted")
    
def update_employee():
    col=input("enter the column name to be updated")
    nw=input("enter the new data")
    idd=int(input("enter the id of the employee to be updated"))
    cursor=mydb.cursor()
    query="UPDATE EMPLOYEE SET %s=%s WHERE EMP_ID=%s"
    value=(col,nw,idd)
    cursor.execute(query,value)
    cursor.fetchall()
    mydb.commit()
    print("....................")
    print("successfully updated")
    
def list_employee():
    cursor=mydb.cursor()
    q="SELECT * FROM `EMPLOYEE`" 
    cursor.execute(q)
    result=cursor.fetchall()
    if result:
        print("employee details")
        print("................")
        for row in result:
            print("employee id:",row[0])
            print("employee name:",row[1])
            print("employee salary:",row[2])
            print("employee department:",row[3])
            print("................")
def delete_employee(a):
    cursor=mydb.cursor()
    query="DELETE FROM EMPLOYEE WHERE EMP_ID=%s"
    d=(a,)
    cursor.execute(query,d)
    cursor.fetchall()
    mydb.commit()
    print("....................")
    print("successfully deleted")
    
def sort_employee():
    
    cursor=mydb.cursor()
    query="SELECT * FROM `EMPLOYEE` ORDER BY EMP_SALARY ASC"
    
    cursor.execute(query)
    result=cursor.fetchall()
    
    for X in result:
        print(X)
            
            
    
    
if __name__=="__main__":
    
    while True:
    
        print("EMPLOYEE MANAGEMENT SYSTEM")
        print("1. search an employee")
        print("2.insert employee details")
        print("3.update employee details")
        print("4.list employees")
        print("5.delete employees")
        print("6.Sorting")
        print("8.exit")
        
        ch=int(input("enter the choice"))
        if ch==1:
            keyword=input("enter the employee name: ")
            searchemployee(keyword)
        
        elif ch==2:
            E_id=int(input("enter the id of the employee"))
            name=input("enter the name of the employee")
            salary=int(input("enter the salary of the employee"))
            dept=input("enter the idepartment  of the employee")
            mail=input("enter the mail id of the employee")
            insert_employee(E_id,name,salary,dept,mail)

        elif ch==3:
             update_employee()
                
        elif ch==4: 
            list_employee()
            
        elif ch==5:
            idd=int(input("enter the id of the employee to be deleted"))
            delete_employee(idd)
        elif ch==6:
            print("sorting on the basis of employee salary  ")
            sort_employee()
              
        elif ch==8:
            break
            
        else:
            print("invalid number,try again")
            print("...........................")
            print("...........................")
    mydb.close()
         
```
